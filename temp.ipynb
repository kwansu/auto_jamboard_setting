```python
import matplotlib.pyplot as plt
import cv2


a = cv2.imread("2.png", cv2.IMREAD_UNCHANGED)
print(a.shape)
plt.imshow(a)
```

```python
b = cv2.imread("3.png", cv2.IMREAD_UNCHANGED)
print(b.shape)
plt.imshow(b)
```
